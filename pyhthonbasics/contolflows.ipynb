```python
#if statement
val=int(input("enter the number"))
if(val%2==0):
    print("number is even",val)
else:
    print("number is odd",val)
```

```python
##loops
list=[1,2,3,4,5,6]
for i in list:
    print(i)
```

```python
even_sum=0
odd_sum=0
for i in list:
    if(i%2==0):
        even_sum+=i
    else:
        odd_sum+=i
print("even sum is {}".format(even_sum),"\nodd sum is {}".format(odd_sum))
```

```python
#while loop
even_sum=0
odd_sum=0
i=0
while(i<=10):
    if(i%2==0):
        even_sum=even_sum+i
    else:
        odd_sum=odd_sum+i
    i=i+1
print("evensum is {} odd aum is {}".format(even_sum,odd_sum))
```

```python
#break condition
n=0
while(n<=10):
    print(n)
    n=n+1
    if(n==5):
        break
```

```python
#continue condition
i=0
while(i<=10):
    i=i+1
    if(i==5):
        continue
    print(i)
```
